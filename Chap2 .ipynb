# for 迴圈

```python
#1+2+3...+10總和
sum = 0
for count in range(1, 11, 1): #從1開始，間隔1, 不包含11
    sum += count
print("1+2+3+...+10 =", sum)
```

```python
#range另外一種寫法
#1+2+3...+10總和
sum = 0
for count in range(11): #1-10, 不包含11
    sum += count
print("1+2+3+...+10 =", sum)
```

# 雞兔同籠，若共有35個頭，94隻腳，請問有幾隻雞?幾隻兔?

```python
for chicken in range(0,36):
    rabbit=35-chicken
    if (2*chicken+4*rabbit) == 94:
        print("雞有",chicken,"隻,兔子有",rabbit,"隻")
        break
else:
    print("無解")
```

# 本金10000元，定存年利率0.6%，請問多久銀行存款會翻倍

# 帳號密碼判斷

```python
id="tony"
pwd="1234"
while True:
    x=input("請輸入帳號：")
    y=input("請輸入密碼：")
    if (x==id and y==pwd):
        print("歡迎登入")
        break
    else:
        print("帳號或密碼輸入錯誤")
```

# 猜數字

```python
import random
number=random.randint(1,20)
while True:
    guess=int(input("請輸入1~20之間的數字-->"))
    if guess == number:
        print("你猜對了，數字是", number)
        break
    else:
        print("你猜錯了")
```
